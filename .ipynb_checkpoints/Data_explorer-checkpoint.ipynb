```python
def set_agegroup(x):
    if x < 50:
        return 0
    elif (x >= 50) & (x < 60):
        return 1
    elif (x > 60):
        return 2

full_df['age_group'] = full_df.apply(lambda full_df: set_agegroup(full_df['age_center.0.0']), axis = 1)

# sns.scatterplot(data = full_df, x = 'tsne_0', y = 'tsne_1', hue = 'sex', s = 1)
```

```python
def set_frailgroup(x):
    if x < 0.2:
        return 0
    elif (x >= 0.2):
        return 1


full_df['frail_group'] = full_df.apply(lambda full_df: set_frailgroup(full_df['FI_0']), axis = 1)
```

## Creating subsets based on age

```python
cmb_train_df = pd.read_csv(prefix + "/full_train_cmb.csv")
cmb_val_df = pd.read_csv(prefix + "/full_val_cmb.csv")

basicinfo = pd.read_csv("Data/basicinfo_instance_0.csv")
```

```python
cmb_train_winfo = cmb_train_df.merge(basicinfo, on = 'eid')
cmb_val_winfo = cmb_val_df.merge(basicinfo, on="eid")

cmb_train_4050 = cmb_train_winfo.loc[cmb_train_winfo['age_center.0.0'] < 50]
cmb_train_5060 = cmb_train_winfo.loc[(cmb_train_winfo['age_center.0.0'] >= 50) & (cmb_train_winfo['age_center.0.0'] < 60)]
cmb_train_6070 = cmb_train_winfo.loc[(cmb_train_winfo['age_center.0.0'] >= 60) & (cmb_train_winfo['age_center.0.0'] <= 70)]

cmb_val_4050 = cmb_val_winfo.loc[cmb_val_winfo['age_center.0.0'] < 50]
cmb_val_5060 = cmb_val_winfo.loc[(cmb_val_winfo['age_center.0.0'] >= 50) & (cmb_val_winfo['age_center.0.0'] < 60)]
cmb_val_6070 = cmb_val_winfo.loc[(cmb_val_winfo['age_center.0.0'] >= 60) & (cmb_val_winfo['age_center.0.0'] <= 70)]


cmb_train_4050.drop(columns = basicinfo.columns.drop('eid'),inplace=True)
cmb_train_5060.drop(columns = basicinfo.columns.drop('eid'),inplace=True)
cmb_train_6070.drop(columns = basicinfo.columns.drop('eid'),inplace=True)

cmb_val_4050.drop(columns = basicinfo.columns.drop('eid'),inplace=True)
cmb_val_5060.drop(columns = basicinfo.columns.drop('eid'),inplace=True)
cmb_val_6070.drop(columns = basicinfo.columns.drop('eid'),inplace=True)
```

```python
cmb_train_4050.to_csv("cmb_train_4050.csv", index = False)
cmb_train_5060.to_csv("cmb_train_5060.csv", index = False)
cmb_train_6070.to_csv("cmb_train_6070.csv", index = False)

cmb_val_4050.to_csv("cmb_set2_4050.csv", index = False)
cmb_val_5060.to_csv("cmb_set2_5060.csv", index = False)
cmb_val_6070.to_csv("cmb_set2_6070.csv", index = False)
```
