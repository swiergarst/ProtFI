```python
import pandas as pd
import statsmodels.api as sm

# Sample DataFrame (replace with your actual DataFrame)
# fatih_comb = pd.read_csv('your_data.csv')

# Define the relevant columns (convert to lower case)
fatih_prots = ["C3", "C8G", "C4B", "C5", "IGFALS", "GSN", "PROS1", "FUCA2", "C2", "SERPINF1", "IGKV3-20", "FN1", "PRG4", "GPLD1", "ERN1", "LYVE1",
               "F5", "ACTB", "ATRN", "PRG4"]
fatih_prots_lower = [col.lower() for col in fatih_prots]

# Identify common columns in fatih_prots and fatih_comb (convert to lower case)
common_columns = [col for col in fatih_prots_lower if col in fatih_comb.columns.str.lower()]

# Initialize an empty list to store results
results_list = []

# Loop through each sex category
for sex in [0, 1]:
    # Filter the DataFrame for the current sex
    subset = fatih_comb[fatih_comb['sex'] == sex]
    
    # Loop through each common column and run the regression
    for col in common_columns:
        # Prepare the design matrix (X) and response variable (y)
        X = subset[[col, 'age_center.0.0']]  # Include age in the design matrix
        y = subset['FI_0']
        
        # Add a constant to the model (intercept)
        X = sm.add_constant(X)
        
        # Fit the regression model
        model = sm.OLS(y, X).fit()
        
        # Create a results dictionary for the current column
        result = {
            'Sex': sex,
            'Variable': col,
            'Coefficient': model.params[col],
            'Standard Error': model.bse[col],
            't-Value': model.tvalues[col],
            'P-Value': model.pvalues[col],
            'LL': model.conf_int().iloc[0, 0],  # Lower limit
            'UL': model.conf_int().iloc[0, 1],  # Upper limit
            'R-squared': model.rsquared,
        }
        
        # Append the result to the results list
        results_list.append(result)

# Create a DataFrame from the results list
results_df = pd.DataFrame(results_list)

# Print the results DataFrame
print(results_df)

# Optionally, save the results to a CSV file
results_df.to_csv('fatih_regression_results_per_sex.csv', index=False)
```
